```python
# importing the  second dataset to add new column
customer_data = pd.read_csv('data/cust.csv')
customer_data.head()
```

```python
Segmentation_df = clean_data[['user_id','gender','age','income']].merge(customer_data[['user_id','Number of orders']], on='user_id', how='left')
```

```python
#check tail of data for nan
Segmentation_df.tail()
```

```python
Segmentation_df.dropna(inplace=True)
```

```python
Segmentation_df.tail()
```

```python
#rename final coulumn 
Segmentation_df.rename(columns={'Number of orders': 'customer score'}, inplace=True)
Segmentation_df.head()
```
