```python
# for basic mathematics operation 
import numpy as np
import pandas as pd
from pandas import plotting

# for visualizations
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')

# for interactive visualizations
import plotly.offline as py
from plotly.offline import init_notebook_mode, iplot
import plotly.graph_objs as go
from plotly import tools
init_notebook_mode(connected = True)
import plotly.figure_factory as ff

# for path
import os
```

```python
# importing the  second dataset to add new column
customer_data = pd.read_csv('data/CustomerOrders.csv')
customer_data.head()
```

```python
#import clean dataset
clean_data = pd.read_csv('data/clean_profile.csv')
clean_data.head()
```

```python
#rename columns to match
customer_data.rename(columns={'ID': 'user_id', 'Income': 'income', 'Age': 'age'}, inplace=True)
customer_data.head()
```

```python
Segmentation_df = clean_data[['user_id','gender',]].merge(customer_data[['user_id', 'age','income','# of Orders']], on='user_id', how='left')
```

```python
#check tail of data for nan
Segmentation_df.tail()
```

```python
Segmentation_df.dropna(inplace=True)
```

```python
Segmentation_df.tail()
```

```python
#rename final coulumn 
Segmentation_df.rename(columns={'# of Orders': 'customer score'}, inplace=True)
Segmentation_df.head()
```

```python
#Save New File
output_file_path = "instacart_customer.csv"
Segmentation_df.to_csv(output_file_path, index=False)
```
