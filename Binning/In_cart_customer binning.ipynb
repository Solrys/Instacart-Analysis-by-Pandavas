```python
import pandas as pd
import numpy as np
import seaborn as sns

sns.set_style('whitegrid')

raw_df = pd.read_csv('in_cart_customer.csv')
raw_df.head()
```

```python
raw_df.tail()
```

```python
raw_df.describe()
```

```python
#bin customer score
bin_labels_5 = ['5', '10', '15', '20', '25']
raw_df['customer score'] = pd.qcut(raw_df['customer score'],
                              q=[0, .2, .4, .6, .8, 1],
                              labels=bin_labels_5)
raw_df.head()
```

```python
#view data for income binns
raw_df.describe()
```

```python
#bin income/110,000 symbolizes 110 and above
bin_labels_income = ['30000', '50000', '70000', '90000', '110,000']
raw_df['income'] = pd.qcut(raw_df['income'],
                              q=[0, .2, .4, .6, .8, 1],
                              labels=bin_labels_income)
raw_df.head()
```

```python
#Save New File
output_file_path = "binned_customer.csv"
raw_df.to_csv(output_file_path, index=False)
```
